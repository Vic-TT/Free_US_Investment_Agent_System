```python
import sys
sys.path.append('src')

from main import run_hedge_fund

# 模拟命令行参数
import argparse

parser = argparse.ArgumentParser(description='Run the hedge fund trading system')
parser.add_argument('--ticker', type=str, required=True, help='Stock ticker symbol')
parser.add_argument('--start-date', type=str, help='Start date (YYYY-MM-DD). Defaults to 3 months before end date')
parser.add_argument('--end-date', type=str, help='End date (YYYY-MM-DD). Defaults to today')
parser.add_argument('--show-reasoning', action='store_true', help='Show reasoning from each agent')
parser.add_argument('--initial-capital', type=float, default=100000.0, help='Initial cash amount (default: 100,000)')
parser.add_argument('--num-of-news', type=int, default=5, help='Number of news articles to analyze for sentiment (default: 5)')

# 手动设置参数值
args = parser.parse_args(['--ticker', 'AAPL', '--start-date', '2024-01-01', '--end-date', '2024-12-31'])

# 配置投资组合
portfolio = {
    "cash": args.initial_capital,
    "stock": 0  # No initial stock position
}

# 调用 main.py 中的函数
result = run_hedge_fund(
    ticker=args.ticker,
    start_date=args.start_date,
    end_date=args.end_date,
    portfolio=portfolio,
    show_reasoning=args.show_reasoning,
    num_of_news=args.num_of_news
)

print("\nFinal Result:")
print(result)
```
